```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
critic_data = pd.read_json("data.json")
critic_data = critic_data[critic_data.metacritic_score != "tbd"]
critic_data['metacritic_score'] = pd.to_numeric(critic_data['metacritic_score'])
```

```python
critic_data = critic_data[critic_data.Shooter == 1]
```

```python
sns.scatterplot(x="critic_score", y="metacritic_score", palette="rainbow", hue="Shooter", 
                data=critic_data)
plt.plot([0, 100], [0, 100], linewidth=2)
```

```python
cs = critic_data['critic_score'].tolist()
ms = critic_data['metacritic_score'].tolist()
cs = np.array(cs).astype(int)
ms = np.array(ms).astype(int)
```

```python
fig, ax =plt.subplots(1,2)
sns.boxplot(y="metacritic_score", data=critic_data,color="b", ax = ax[0])
sns.boxplot(y="critic_score", data=critic_data,color="r", ax = ax[1])
```

```python
ttest = stats.ttest_rel(cs,ms)
```

```python
# p-value
ttest.pvalue/2
```
